```python

##pca
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()

X = pd.DataFrame(iris.data)
y = iris.target
```

```python
X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled.shape
```

```python
# PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=2)#want 2 from 4

X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=y
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA for iris dataset")
```

```python
print("explained variance ratio: ", pca.explained_variance_ratio_)
```

```python
print(pca.components_)#consist of 4 feature
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("thyroid_dataset.csv")
df
```

```python
X = df.drop("Outlier_label", axis=1)
y = df["Outlier_label"]# as we are going to predict we will be dropping this output
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.ensemble import IsolationForest#auto

clf = IsolationForest(n_estimators=200, contamination='auto', random_state=42)
```

```python
labels = clf.fit_predict(X_scaled)
labels
```

```python
#reduaction in 2d
# Visualize
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
plt.xlabel("PC1")
plt.ylabel("PC2")
```

```python
import numpy as np

n_outliers = np.sum(labels == -1)
n_normal = np.sum(labels == 1)

print("outlier = ", n_outliers)
print("normal = ", n_normal)
```

```python

clf2= IsolationForest(n_estimators=200, contamination=0.036, random_state=42)
labels = clf2.fit_predict(X_scaled)

plt.figure(figsize=(8, 6))

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
plt.xlabel("PC1")
plt.ylabel("PC2")

```

```python
import numpy as np

n_outliers = np.sum(labels == -1)
n_normal = np.sum(labels == 1)

print("outlier = ", n_outliers)
print("normal = ", n_normal)
```

```python
#LOF
# sometimes isolation forest may not detect local outlier it may consider into the clusters but for the local outlier better ways to use lOF
```

```python
from sklearn.neighbors import LocalOutlierFactor

```

```python
neighbs = LocalOutlierFactor()

labels = neighbs.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
plt.xlabel("PC1")
plt.ylabel("PC2")
```

```python
n_outliers = np.sum(labels == -1)
n_normal = np.sum(labels == 1)

print("outlier = ", n_outliers)
print("normal = ", n_normal)
```

```python
neighbs = LocalOutlierFactor(contamination=0.036)

labels = neighbs.fit_predict(X_scaled)
plt.figure(figsize=(8, 6))

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
plt.xlabel("PC1")
plt.ylabel("PC2")

n_outliers = np.sum(labels == -1)
n_normal = np.sum(labels == 1)

print("outlier = ", n_outliers)
print("normal = ", n_normal)
```
